````python
# ============================================================================
# PRESENTATION SETUP
# ============================================================================
"""
Initialize notebook for presentation mode
"""

# Auto-reload for development
%load_ext autoreload
%autoreload 2

# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import Markdown, display

# Presentation-friendly plotting
plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 14
plt.rcParams["axes.titlesize"] = 18

print("âœ… Presentation mode activated!")
````

---

## ğŸ“ˆ Sample Data Analysis

Let's analyze species distribution data:

````python
# Generate sample data for demonstration
np.random.seed(42)

# Sample species data
species_data = pd.DataFrame(
    {
        "Species": [
            "Rosa canina",
            "Pinus sylvestris",
            "Quercus robur",
            "Betula pendula",
            "Alnus glutinosa",
        ],
        "Occurrences": np.random.randint(1000, 10000, 5),
        "Countries": np.random.randint(10, 50, 5),
        "Data_Quality": np.random.uniform(0.7, 0.95, 5),
    }
)

# Display data
display(Markdown("### ğŸ“Š Species Dataset"))
display(
    species_data.style.background_gradient(
        cmap="Greens", subset=["Occurrences"]
    ).format({"Data_Quality": "{:.2%}"})
)
````

---

## ğŸ“Š Visualization: Species Occurrences

````python
# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Bar chart - Occurrences
species_data.plot(
    x="Species",
    y="Occurrences",
    kind="bar",
    ax=ax1,
    color="forestgreen",
    alpha=0.7,
    legend=False,
)
ax1.set_title("Species Occurrence Records", fontsize=18, fontweight="bold")
ax1.set_xlabel("Species", fontsize=14)
ax1.set_ylabel("Number of Records", fontsize=14)
ax1.tick_params(axis="x", rotation=45)
ax1.grid(axis="y", alpha=0.3)

# Scatter plot - Quality vs Coverage
scatter = ax2.scatter(
    species_data["Countries"],
    species_data["Data_Quality"],
    s=species_data["Occurrences"] / 10,
    c=species_data["Occurrences"],
    cmap="Greens",
    alpha=0.7,
    edgecolors="black",
)
ax2.set_title("Data Quality vs Geographic Coverage", fontsize=18, fontweight="bold")
ax2.set_xlabel("Number of Countries", fontsize=14)
ax2.set_ylabel("Data Quality Score", fontsize=14)
ax2.grid(alpha=0.3)

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax2)
cbar.set_label("Occurrences", fontsize=12)

plt.tight_layout()
plt.show()

print("ğŸ“Š Visualization complete!")
````
