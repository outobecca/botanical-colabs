````python
# ============================================================================
# WRITEFILE MAGIC EXAMPLE (Best Practice #2)
# ============================================================================
"""
This cell demonstrates %%writefile magic for code modularization.

Benefits:
- Keep code organized in separate files
- Test functions in notebook context
- Easy to reuse in other notebooks
- Version control for utility functions

The file will be created and can be imported with auto-reload!
"""

# First, create the utils directory if it doesn't exist
import os
os.makedirs('utils', exist_ok=True)
print("ğŸ“ Created utils/ directory")
````

````python
%%writefile utils/api_helpers.py
"""
API Helper Functions for Botanical Data Retrieval

This module is automatically written by the notebook using %%writefile magic.
Edit this cell and re-run to update the file!

Usage:
    from utils.api_helpers import safe_api_call, validate_species_name
"""

from typing import Dict, Optional, Any
import requests


def safe_api_call(url: str, 
                  params: Optional[Dict[str, Any]] = None,
                  timeout: int = 15,
                  method: str = 'GET') -> Optional[Dict[str, Any]]:
    """
    Safe API call with comprehensive error handling.
    
    Args:
        url: API endpoint URL
        params: Query parameters or JSON body
        timeout: Timeout in seconds
        method: HTTP method (GET or POST)
    
    Returns:
        JSON response as dict or None on error
    
    Examples:
        >>> data = safe_api_call("https://api.gbif.org/v1/species/search", 
        ...                      params={"q": "Rosa canina"})
    """
    try:
        if method.upper() == 'GET':
            response = requests.get(url, params=params, timeout=timeout)
        elif method.upper() == 'POST':
            response = requests.post(url, json=params, timeout=timeout)
        else:
            raise ValueError(f"Unsupported method: {method}")
        
        response.raise_for_status()
        return response.json()
        
    except requests.exceptions.Timeout:
        print(f"â±ï¸ Timeout: {url}")
        return None
    except requests.exceptions.HTTPError as e:
        print(f"ğŸš« HTTP {e.response.status_code}: {url}")
        return None
    except requests.exceptions.RequestException as e:
        print(f"âŒ Network error: {str(e)[:100]}")
        return None
    except ValueError as e:
        print(f"âŒ Invalid JSON response from {url}")
        return None


def validate_species_name(name: str) -> bool:
    """
    Validates botanical species name format.
    
    Args:
        name: Species name to validate
    
    Returns:
        True if valid format, False otherwise
    
    Examples:
        >>> validate_species_name("Rosa canina")
        True
        >>> validate_species_name("x")
        False
    """
    if not name or len(name) < 3:
        print("âš ï¸ Species name too short")
        return False
    
    # Basic validation: should contain at least one space (genus + species)
    if ' ' not in name.strip():
        print("â„¹ï¸ Note: Single-word names are unusual for species")
    
    return True


def format_scientific_name(name: str) -> str:
    """
    Formats species name in proper scientific notation.
    
    Args:
        name: Species name
    
    Returns:
        Properly formatted scientific name in italics (HTML)
    
    Examples:
        >>> format_scientific_name("rosa canina")
        '<i>Rosa canina</i>'
    """
    words = name.strip().split()
    if len(words) >= 2:
        # Capitalize genus, lowercase species
        formatted = f"{words[0].capitalize()} {words[1].lower()}"
        # Add subspecies/variety if present
        if len(words) > 2:
            formatted += " " + " ".join(words[2:]).lower()
        return f"<i>{formatted}</i>"
    return name


print("âœ… Module written to utils/api_helpers.py")
````

### API Endpoints

```{list-table}
:header-rows: 1

* - Service
  - Endpoint
  - Documentation
* - GBIF
  - `api.gbif.org/v1/species`
  - [GBIF API Docs](https://www.gbif.org/developer/species)
* - Wikipedia
  - `en.wikipedia.org/api/rest_v1`
  - [Wikipedia API](https://www.mediawiki.org/wiki/API:Main_page)
```

````python
# ============================================================================
# DATA FETCHING FUNCTIONS
# ============================================================================
"""
Functions for retrieving botanical data from multiple sources
"""

def fetch_gbif_data(species_name: str) -> Optional[Dict[str, Any]]:
    """
    Fetches taxonomy and distribution from GBIF.
    
    Args:
        species_name: Scientific name
    
    Returns:
        Dictionary with GBIF data or None
    
    References:
        GBIF.org (2025) GBIF Occurrence Download 
        https://doi.org/10.15468/dl.XXXXX
    """
    url = "https://api.gbif.org/v1/species/search"
    params = {"q": species_name, "limit": 1}
    
    data = safe_api_call(url, params)
    
    if not data or not data.get('results'):
        print(f"âŒ No GBIF data for: {species_name}")
        return None
    
    result = data['results'][0]
    
    return {
        "scientificName": result.get('scientificName', 'N/A'),
        "kingdom": result.get('kingdom', 'N/A'),
        "family": result.get('family', 'N/A'),
        "genus": result.get('genus', 'N/A'),
        "species": result.get('species', 'N/A'),
        "taxonomicStatus": result.get('taxonomicStatus', 'N/A'),
        "gbifKey": result.get('key', 'N/A')
    }


def fetch_wikipedia_summary(species_name: str, language: str = 'en') -> Optional[str]:
    """
    Fetches Wikipedia summary in specified language.
    
    Args:
        species_name: Scientific name
        language: Language code (en, fi, sv, etc.)
    
    Returns:
        Summary text or None
    """
    url = f"https://{language}.wikipedia.org/api/rest_v1/page/summary/{species_name.replace(' ', '_')}"
    
    data = safe_api_call(url)
    
    if data and 'extract' in data:
        return data['extract']
    
    return None


print("âœ… Data fetching functions defined")
````

(analysis-section)=
## ğŸ”¬ Data Analysis

:::{important}
Ensure configuration is saved before running analysis
:::

````python
# ============================================================================
# MAIN ANALYSIS EXECUTION
# ============================================================================
"""
Executes complete data collection and analysis workflow
"""

if not CONFIG["species_name"]:
    print("âŒ ERROR: Enter species name in configuration panel")
else:
    species = CONFIG["species_name"]
    lang = CONFIG["language"]
    
    print(f"\n{'='*60}")
    print(f"ğŸŒ¿ Analyzing: {format_scientific_name(species)}")
    print(f"ğŸŒ Language: {lang}")
    print(f"{'='*60}\n")
    
    # Validate input
    if not validate_species_name(species):
        print("âš ï¸ Continuing with potentially invalid species name...\n")
    
    # Collect data
    all_data = {}
    
    # 1. GBIF Data
    print("ğŸ“Š Fetching GBIF taxonomy...")
    gbif_data = fetch_gbif_data(species)
    if gbif_data:
        all_data['gbif'] = gbif_data
        print(f"   âœ… Found: {gbif_data.get('scientificName')}")
    
    # 2. Wikipedia Summary
    print(f"\nğŸ“š Fetching Wikipedia summary ({lang})...")
    wiki_summary = fetch_wikipedia_summary(species, lang)
    if wiki_summary:
        all_data['wikipedia'] = wiki_summary
        print(f"   âœ… Retrieved {len(wiki_summary)} characters")
    
    print(f"\n{'='*60}")
    print("âœ… Data collection complete!")
    print(f"{'='*60}")
````

````python
# ============================================================================
# GENERATE CITATIONS
# ============================================================================
"""
Creates properly formatted citations for all data sources used
"""

from datetime import datetime

def generate_citation(source: str, date: str = None) -> str:
    """Generates citation for data source"""
    if date is None:
        date = datetime.now().strftime('%Y-%m-%d')
    
    citations = {
        "gbif": f"GBIF.org ({date.split('-')[0]}) GBIF Occurrence Download. https://doi.org/10.15468/dl.XXXXX",
        "wikipedia": f"Wikipedia contributors. ({date.split('-')[0]}). Retrieved {date}, from https://en.wikipedia.org",
        "trefle": f"Trefle.io ({date.split('-')[0]}). Global Plants Database. https://trefle.io"
    }
    
    return citations.get(source, f"{source} ({date})")

# Display citations
if 'all_data' in globals() and all_data:
    display(Markdown(f"\n## ğŸ“š Data Source Citations"))
    display(Markdown(f"*Data retrieved: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}*\n"))
    
    for source in all_data.keys():
        citation = generate_citation(source)
        display(Markdown(f"**{source.upper()}:** {citation}"))

print("\nâœ… Citations generated")
````
